Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
# Step 1 - preparing and claening data in anylized columns
df_age_language = df_results[['ResponseId', 'Age', 'LanguageHaveWorkedWith']].dropna(subset = ['Age', 'LanguageHaveWorkedWith'])
df_age_language
```

```python
# Step 2 - Grouping by age - all responders
age_all = df_age_language.groupby('Age')['ResponseId'].count() # or .size()
age_all 
```

```python
# Step 3 - Grouping by age -  responders, working with Python
age_py = df_age_language[df_age_language['LanguageHaveWorkedWith'].str.contains('python', case = False, na = False)].groupby('Age')['ResponseId'].count()
age_py
```

```python
# Step 4 - final df + percentage
df_py_popularity_by_Age = age_all.to_frame(name='Total_Resp')
df_py_popularity_by_Age['Python_Resp'] = age_py

df_py_popularity_by_Age['Py_percent'] = (df_py_popularity_by_Age['Python_Resp']/df_py_popularity_by_Age['Total_Resp'] * 100).round(2)
display(df_py_popularity_by_Age.reset_index())

#OR
Py_percent = (df_py_popularity_by_Age['Python_Resp']/df_py_popularity_by_Age['Total_Resp'] * 100).round(2).to_frame(name='Python_pct').reset_index()
display(Py_percent)
```
